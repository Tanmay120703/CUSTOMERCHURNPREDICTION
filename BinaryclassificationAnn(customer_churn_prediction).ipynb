```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df2 = df.copy()
```

```python
df2.drop(columns = ['RowNumber' , 'Surname', 'CustomerId'] , inplace = True)
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
x = df2.iloc[:,:10]
y = df2.iloc[:,-1]
```

```python
x
```

```python
x_train ,  x_test, y_train , y_test = train_test_split(x , y, test_size=0.2 , random_state=42)
```

```python
categorical_columns = ['Geography' , 'Gender']
```

```python
numerical_columns = ['CreditScore' , 'Age' , 'Tenure' , 'Balance' , 'NumOfProducts' , 'EstimatedSalary']
```

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(
    transformers=[
    ('OHE' , OneHotEncoder(sparse_output=False , drop = 'first') , categorical_columns),
    ('pass' , StandardScaler(), numerical_columns)
])
```

```python
x_train_transformed = transformer.fit_transform(x_train)
```

```python
x_test_transformed = transformer.transform(x_test)
```

```python
class customdataset(Dataset):
  def __init__(self , features , labels):
    self.features = torch.tensor(features , dtype = torch.float32)
    self.labels = torch.tensor(labels , dtype = torch.float32)

  def __len__(self):
    return len(self.features)

  def __getitem__(self,index):
    return self.features[index] , self.labels[index]
```

```python
train_dataset = customdataset(x_train_transformed , y_train.values)
test_dataset = customdataset(x_test_transformed , y_test.values)
```

```python
train_dataloader = DataLoader(train_dataset , shuffle = True , batch_size = 40)
test_dataloader = DataLoader(test_dataset , shuffle = True , batch_size = 40)
```

```python
len(train_dataset)
```

```python
import torch.nn as nn
```

```python
class myann(nn.Module):
  def __init__(self , num_features):
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(num_features , 64 ),
        nn.ReLU(),
        nn.Linear(64 , 32),
        nn.ReLU(),
        nn.Linear(32,1)
    )

  def forward(self , num_features):
    return self.model(num_features)
```

```python
model1 = myann(x_train_transformed.shape[1])
epochs = 100
learning_rate = 0.1
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model1.parameters() , lr = learning_rate)
```

```python
for epoch in range(epochs):
  loss1 = 0
  for batch_features , batch_labels in train_dataloader:
    output = model1(batch_features)

    batch_labels = batch_labels.view(-1, 1).float()

    loss = criterion(output , batch_labels)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

  loss1 += loss.item()
  print(loss1)
```
